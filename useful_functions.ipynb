```python
def sep(file):
    #path='../new_csv/'+file
    
    path=file
    f=pd.read_csv(path)
    frame=f
    column=frame.columns.values[2]
    frame=frame.loc[frame[column]!='None'].reset_index()
    result=frame
    print(frame.iloc[1].values)
    if type(frame.loc[1,column])==str:
        for i in range(len(frame)):
            frame.loc[i,column]=frame.loc[i,column].strip('{').strip('}')
        sf=frame[column].str.split(pat=',',expand=True).stack()
        sf=sf.reset_index(level=1,drop=True).rename(column)
        b=result['business_id']
        result=pd.concat([sf,b],axis=1)
     
    result['relation']=column
    result=result[['business_id','relation',column]]
    
    if type(frame.loc[1,column])==str:
        if 'True' in result.loc[1,column] or 'False' in result.loc[1,column]:
            result=result.loc[result[column].str.contains('True')]
            result[column]=column.strip('attributes.')
    elif type(frame.loc[1,column])== numpy.bool_:
        result=result.loc[result[column]==True]
        result[column]=column.strip('attributes.')
             #if type(frame.loc[1,column])==str:
    #else:
        #result=result.loc[result[column].str.contains('True')]
    result.to_csv('../new_kg/'+column+'.csv')
    return result
```

```python
def assig_index(file,relationdict,itemdict,businessdict,ri,ii):
    df=pd.read_csv(file)
    attributes=df.columns.values[-1]
    #print(attributes)
    df=df[['business_id','relation',attributes]]

    itemtxt=open('../6/item_index.txt','a',encoding='utf-8')
    relationtxt=open('../6/relation_index.txt','a',encoding='utf-8')
    businesstxt=open('../6/business_index.txt','a',encoding='utf-8')
    for i in range(len(df)):
        business_id=df.loc[i,'business_id'].strip()
        relation=df.loc[i,'relation'].strip()
        attributes_value=str(df.loc[i,attributes]).strip()
        
        if business_id not in itemdict:
            
            businessdict[business_id]=ii
            businesstxt.write('%s\t%d\n'%(business_id,itemdict[business_id]))
            df.loc[i,'business_id']=ii
            ii+=1
            
        elif business_id in itemdict:
            df.loc[i,'business_id']=businessdict[df.loc[i,'business_id']]
            
        if relation not in relationdict:
            
            
            relationdict[relation]=ri
            relationtxt.write(('%s\t%d\n'%(relation,relationdict[relation])))
            #print(('%s\t%d\n'%(relation,relationdict[relation])))
            df.loc[i,'relation']=relationdict[relation]
            ri+=1
        elif relation in relationdict:
            df.loc[i,'relation']=relationdict[df.loc[i,'relation']]
            
        if attributes_value not in itemdict:
            
            itemdict[attributes_value]=ii
            itemtxt.write('%s\t%d\n'%(attributes_value,itemdict[attributes_value]))
            
           # print(attributes_value,itemdict[attributes_value])
            df.loc[i,attributes]=ii
            ii+=1
       
        elif attributes_value in itemdict:
            df.loc[i,attributes]=itemdict[attributes_value]
    itemtxt.close()
    relationtxt.close()
    print(ri,ii)
    df.to_csv('../6/'+attributes+'.csv')
    output=[relationdict,itemdict,ri,ii]
    return output
```

```python
def reorder(path):
    df=pd.read_csv(path)
    colname=df.columns.values
    print(colname)
    third=len(colname)-1
    firstcol=colname[third-2]
    seccol=colname[third-1]
    thirdcol=colname[third]
    #if colname[2]!='relation':
       # print('error')
    #else:
    df=df[[firstcol,seccol,thirdcol]]
    df=df.rename({'business_id':'item',thirdcol:'item.2'},axis='columns')
    df.to_csv('../new_kg2/'+thirdcol+'.csv')#output the file name as the third column's name.
    
```
